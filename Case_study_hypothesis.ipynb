Case Study on Testing of Hypothesis

2 a. The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing.

```python
import pandas as pd

df = pd.read_csv("/content/Sales_add.csv")
```

```python
df.info()
```

```python
df
```

```python
from statsmodels.stats.weightstats import ztest

z_stat,p_value = ztest(df['Sales_After_digital_add(in $)'],value=100)
```

```python
z_stat,p_value
```

```python
if p_value < 0.05 :
  print ("Reject Null Hypothesis")
else :
  print("Accept Null Hypothesis")
```

Conclusion :there is increase in sales after stepping into digital marketing
